MANIPULANDO DADOS DO EXCEL COM PYTHON

```python
#pandas

import pandas as pd

tabela = pd.read_excel("Produtos.xlsx")
display(tabela)
```

```python
#atualizar a comissão de Produtos
tabela.loc[tabela["Tipo"]=="Produto", "Comissão %"] = 3.0

#calcular o valor da comissão
tabela["Valor Comissão"] = tabela["Comissão %"] * tabela["Preço Base Original"] /100

#mostrar a tabela atualizada
display(tabela)
```
